```python
# import libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
# load the dateset
df1 = pd.read_csv('/content/medical_clean.csv')
```

```python
# create a new dataset with only selected data
df = df1[['ReAdmis','HighBlood','Stroke','Overweight','Arthritis','Diabetes','Hyperlipidemia','BackPain','Anxiety','Reflux_esophagitis','Asthma']]
df
```

# Chi-Square

```python
# ReAdmis and HighBlood Contingency table
contingency_HighBlood = pd.crosstab(df['ReAdmis'], df['HighBlood'])
print(contingency_HighBlood)
```

```python
stat, p_value, dof, expected_value = chi2_contingency(contingency_HighBlood)
prob = 0.95
alpha = 1 - prob
critical_value = chi2.ppf(prob,dof)
print("Chi-Square Statistic =", stat)
print("p-value =", p_value)
print("Degrees of Freedom =", dof)
print('Significance level =%.3f' % (1-prob))
print("Critical value =" , critical_value)
print("Expected value =", expected_value)
```

# Univariate Statistics

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_s = df1[['Initial_days','VitD_levels','ReAdmis','Item2','Item6']]
df_s
```

```python
# rename Item2, Item3 and Item6
df_s.rename(columns = {'Item2':'Timely_treatment',
                       'Item6':'Hours_of_treatment'},
            inplace=True)
df_u=df_s[['Initial_days','VitD_levels','Timely_treatment','Hours_of_treatment']]
df_u
```

```python
# count of values
df_u.count()
```

```python
# number of unique values
df_u.nunique()
```

```python
# datatype
df_u.dtypes
```

```python
df_u.describe()
```

```python
# histogram for continuous Variables
fig,axs = plt.subplots(1,2,figsize=(10, 5))
sns.histplot(df_u.Initial_days,kde=True,color="skyblue",ax=axs[0])
sns.histplot(df_u.VitD_levels,kde=True, color="olive",ax=axs[1])
fig.tight_layout()
```

```python
# countplot for two Categorical Variables
fig,axs = plt.subplots(1,2,figsize=(10, 5))
sns.countplot(df_u.Timely_treatment,ax=axs[0])
sns.countplot(df_u.Hours_of_treatment,ax=axs[1])
fig.tight_layout()
```

# Bivariate statistics

```python
df_b = df_s[['Initial_days','VitD_levels','ReAdmis','Hours_of_treatment']]
df_b
```

```python
# Regression line with scatterplot for two continuous variables
sns.regplot(data=df_b, x='Initial_days', y='VitD_levels', line_kws={'color': 'g'})

```

```python
# barplot for two categorical variables
sns.barplot(data=df_b, y = 'Hours_of_treatment', x='ReAdmis')
```

```python
#  box plot, violin plot, strip plot,kdeplot for continuous variable and categorical variable
fig,axs = plt.subplots(2,2,figsize=(10, 8))
sns.boxplot(x="ReAdmis", y="Initial_days", data=df_b,ax=axs[0,0])
sns.violinplot(x="ReAdmis", y="Initial_days", data=df_b,ax=axs[0,1])
sns.stripplot(x="ReAdmis", y="Initial_days", data=df_b,ax=axs[1,0])
sns.kdeplot(data=df_b,x='Initial_days', hue='ReAdmis', fill = True,ax=axs[1,1])
fig.tight_layout()
```

```python
fig,axs = plt.subplots(2,2,figsize=(10, 8))
sns.boxplot(x="ReAdmis", y="VitD_levels", data=df_b,ax=axs[0,0])
sns.violinplot(x="ReAdmis", y="VitD_levels", data=df_b,ax=axs[0,1])
sns.stripplot(x="ReAdmis", y="VitD_levels", data=df_b,ax=axs[1,0])
sns.kdeplot(data=df_b,x='VitD_levels', hue='ReAdmis', fill = True,ax=axs[1,1])
fig.tight_layout()
```
